```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('dataset/titanic.csv')
df
```

```python
sns.countplot(data=df,x=df['Pclass'])
plt.show()
```

```python
sns.countplot(data=df,x=df['Survived'])
```

```python
sns.countplot(data=df,x=df['Sex'])
```

```python
df['Sex'].value_counts().plot(kind = 'pie',autopct='%1.1f%%')
plt.show()
```

```python
df['Embarked'].value_counts().plot(kind = 'pie',autopct='%1.1f%%')
plt.show()
```

```python
sns.distplot(df,x=df['Age'],kde=True)
plt.show()
```

```python
sns.histplot(df,x=df['Age'])
plt.show()
```

```python
sns.displot(data=df,x=df['Age'],kind='kde')
plt.show()
```
